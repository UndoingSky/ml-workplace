```python
import numpy as np

data_set = np.random.randint(1, 101, size=(20, 4))

student_average = np.mean(data_set, axis=1)
print("Student average: \n", student_average)
print("average subject score :\n" ,np.mean(data_set, axis=0) )

print("Best Student :\n", np.argmax(student_average))



print("Subject 1 average :\n", np.mean(data_set[:, 0]))
print("Subject 2 average :\n", np.mean(data_set[:, 1]))
print("Subject 3 average :\n", np.mean(data_set[:, 2]))
print("Subject 4 average :\n", np.mean(data_set[:, 3]))

print("Subject 1 highest :\n", np.max(data_set[:, 0]))
print("Subject 2 highest :\n", np.max(data_set[:, 1]))
print("Subject 3 highest :\n", np.max(data_set[:, 2]))
print("Subject 4 highest :\n", np.max(data_set[:, 3]))

print("Subject 1 lowest :\n", np.min(data_set[:, 0]))
print("Subject 2 lowest :\n", np.min(data_set[:, 1]))
print("Subject 3 lowest :\n", np.min(data_set[:, 2]))
print("Subject 4 lowest :\n", np.min(data_set[:, 3]))

reshape = data_set.reshape(10 , 8)
print("reshaped_data_set: \n", reshape)

split = np.split(data_set, 2)

classroom_1 = split[0]
classroom_2 = split[1]

merged_classroom = np.vstack((classroom_1, classroom_2))

print(data_set == merged_classroom) #returns true

new_student = np.random.randint(0, 101, size=4)[np.newaxis, :]

updated_dataset = np.append(data_set, new_student, axis=0)
print(updated_dataset)

delete = np.delete(updated_dataset, -1, axis=0) #got expled on the first day xd

insert_at = np.insert(updated_dataset, 5, np.random.randint(0, 101, size=4), axis=0)
print(insert_at)






```
