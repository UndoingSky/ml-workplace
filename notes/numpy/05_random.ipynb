# Random Numbers

NumPy's random utilities: seeding, sampling, distributions, and shuffling.

```python
import numpy as np
np.random.seed(42)

print('Random float in [0, 1):', np.random.rand())
print('Random integers from 1 to 9:', np.random.randint(1, 10))
print('Random sample of 5 values:', np.random.randint(0, 100, size=5))

items = np.array([10, 20, 30, 40, 50])
print('Random choice from array:', np.random.choice(items, size=3, replace=False))
shuffled_items = items.copy()
np.random.shuffle(shuffled_items)
print('Shuffled array:', shuffled_items)
print('Random permutation:', np.random.permutation(items))

print('Random numbers with rand(2,3):', np.random.rand(2,3))
print('Standard normal with randn(5):', np.random.randn(5))
print('Uniform(10,20):', np.random.uniform(10,20,size=5))
```
