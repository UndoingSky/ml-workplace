# Concatenation and Stacking

How to join arrays using `np.concatenate` and the importance of the `axis` parameter.

Common pitfall: pass the arrays as a sequence (tuple/list).

- Incorrect: `np.concatenate(a, b)`
- Correct: `np.concatenate((a, b))`

```python
import numpy as np

# Joining 1D arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
joined = np.concatenate((a, b))  # note the tuple
print('Joined 1D:', joined)
```

axis parameter (important for 2D+ arrays):

- `axis=0` (default): stack along the first axis (rows) — vertical stacking.
- `axis=1`: stack along the second axis (columns) — horizontal stacking.
- `axis=None`: flattens inputs then concatenates into 1D.

```python
matrix1 = np.array([[1, 2],
                    [3, 4]])
matrix2 = np.array([[5, 6],
                    [7, 8]])

# Vertical (axis=0)
vert = np.concatenate((matrix1, matrix2), axis=0)
print('Vertical (axis=0):
', vert)

# Horizontal (axis=1)
horiz = np.concatenate((matrix1, matrix2), axis=1)
print('
Horizontal (axis=1):
', horiz)

# Flatten then concatenate (axis=None)
flat_concat = np.concatenate((matrix1, matrix2), axis=None)
print('
Flatten then concatenate (axis=None):', flat_concat)
```

Notes:

- For convenience, NumPy also offers `np.vstack`, `np.hstack`, and `np.stack` for common stacking patterns.
- Ensure shapes are compatible for the chosen axis.
