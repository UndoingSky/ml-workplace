# Array Modification: split, append, insert, delete

Examples and notes on `np.split`, `np.append`, `np.insert`, and `np.delete`.

## `np.split`

- Splits an array into multiple sub-arrays. Use `indices_or_sections` to control splitting.

```python
import numpy as np
arr = np.arange(8)
# Split into 4 equal parts
parts = np.split(arr, 4)
print('parts:', parts)
# Split at indices -> produces uneven sections
parts2 = np.split(arr, [2, 5])
print('parts2:', parts2)
```

## `np.append`

- Returns a new array with values appended to the end. For multi-dimensional arrays, use `axis` to append along a specific axis.

```python
# 1D append
a = np.array([1, 2, 3])
b = np.array([4, 5])
app = np.append(a, b)
print('append 1D:', app)

# 2D append along rows (axis=0)
m1 = np.array([[1,2],[3,4]])
m2 = np.array([[5,6]])
app2 = np.append(m1, m2, axis=0)
print('append 2D axis=0:', app2)
```

## `np.insert`

- Inserts values before the given indices. Returns a new array; for multi-d arrays supply `axis`.

```python
arr = np.array([1, 2, 3, 4])
# Insert 99 before index 2
ins = np.insert(arr, 2, 99)
print('insert:', ins)
# Insert multiple values at positions
ins2 = np.insert(arr, [1,3], [9, 10])
print('insert multiple:', ins2)

# 2D insert along axis=1 (columns)
m = np.array([[1,2],[3,4]])
m_ins = np.insert(m, 1, [99, 100], axis=1)
print('2D insert axis=1:', m_ins)
```

## `np.delete`

- Deletes sub-arrays along an axis and returns a new array. Indices can be a slice or a list of indices.

## Notes and gotchas

- Most of these functions return a new array (they do not modify in-place).
- `np.append` is not the same as `list.append` — it returns a new array (and can be slow in loops).
- Use `axis` carefully for multi-dimensional arrays to ensure shapes match.
- For large-scale insertion/deletion, consider building lists and converting to arrays, or using specialized data structures.

```python
import numpy as np

arr = np.array([1,2,3,4,5])
# Delete index 1
d = np.delete(arr, 1)
print('delete index 1:', d)
# Delete multiple indices
d2 = np.delete(arr, [0,3])
print('delete multiple:', d2)

# 2D delete along axis=0
m = np.array([[1,2],[3,4],[5,6]])
m_del = np.delete(m, 1, axis=0)
print('2D delete axis=0:', m_del)
```
