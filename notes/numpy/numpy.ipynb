```python
import numpy as np

# NumPy basics quickstart

# 1) Create arrays
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2, 3], [4, 5, 6]])

print("a:\n", a)
print("b:\n", b)

# 2) Array properties
print("shape:", b.shape)
print("ndim:", b.ndim)
print("size:", b.size)
print("dtype:", b.dtype)
```

```python
# Simple note: a multidimensional array has more than one axis (dimension).
# In NumPy, each axis represents a level (e.g., rows and columns in 2D arrays).

print("Multidimensional array note:")
print(f"`b` is {b.ndim}D with shape {b.shape} and total elements {b.size}.")
```

```python
# Slicing with indexing note:
# Indexing selects a single element, while slicing selects a range of elements.

print("Slicing with indexing note:")
print("Indexing example: b[0, 1] ->", b[0, 1])
print("Slicing example: b[:, 1] ->", b[:, 1])
print("Slicing example: a[1:3] ->", a[1:3])
```

```python
# Arithmetic operations in arrays note:
# NumPy supports element-wise arithmetic operations between arrays and scalars.
# Operations are performed element-by-element on arrays of the same shape.

print("Arithmetic operations in arrays note:")
print("Addition: a + 2 ->", a + 2)
print("Subtraction: a - 1 ->", a - 1)
print("Multiplication: a * 3 ->", a * 3)
print("Division: a / 2 ->", a / 2)
print("Element-wise operations: b * 2 ->\n", b * 2)
print("Array addition: a + a ->", a + a)
print("Array multiplication (element-wise): b * b ->\n", b * b)
```

```python
# Square root and rounding functions note:
# NumPy provides sqrt() for element-wise square root and round() for rounding values.
# These functions operate element-by-element on arrays.

print("Square root and rounding functions note:")
print("Square root: np.sqrt(a) ->", np.sqrt(a))
print("Square root: np.sqrt(b) ->\n", np.sqrt(b))
print("Rounding: np.round(np.sqrt(a), 2) ->", np.round(np.sqrt(a), 2))
print("Rounding: np.round(a / 2, 1) ->", np.round(a / 2, 1))
```

```python
# Pi function note:
# NumPy provides np.pi, a mathematical constant representing the value of pi (π ≈ 3.14159...).
# It can be used in calculations involving circles, trigonometry, and other mathematical operations.

print("Pi function note:")
print("Value of pi: np.pi ->", np.pi)
print("Circle area with radius 2: np.pi * 2**2 ->", np.pi * 2**2)
print("Circle circumference with radius 2: 2 * np.pi * 2 ->", 2 * np.pi * 2)
```

```python
# Aggregation functions note:
# NumPy provides common summary statistics such as sum(), mean(), std(), and var().
# These functions can be applied to arrays to compute overall or axis-wise statistics.

print("Aggregation functions note:")
print("Sum: np.sum(a) ->", np.sum(a))
print("Mean: np.mean(a) ->", np.mean(a))
print("Standard deviation: np.std(a) ->", np.std(a))
print("Variance: np.var(a) ->", np.var(a))

print("Sum of b: np.sum(b) ->", np.sum(b))
print("Mean of b: np.mean(b) ->", np.mean(b))
print("Standard deviation of b: np.std(b) ->", np.std(b))
print("Variance of b: np.var(b) ->", np.var(b))
```

```python
# Min and max functions note:
print("Min and max functions note:")
print("Min of a: np.min(a) ->", np.min(a))
print("Max of a: np.max(a) ->", np.max(a))
print("Min of b: np.min(b) ->", np.min(b))
print("Max of b: np.max(b) ->", np.max(b))
print("Min of b along axis 0 (columns): np.min(b, axis=0) ->", np.min(b, axis=0))
print("Max of b along axis 0 (columns): np.max(b, axis=0) ->", np.max(b, axis=0))
print("Min of b along axis 1 (rows): np.min(b, axis=1) ->", np.min(b, axis=1))
print("Max of b along axis 1 (rows): np.max(b, axis=1) ->", np.max(b, axis=1))
```

```python
# Random numbers note:
# NumPy's random module can generate random values for simulations, sampling, and testing.
# Use a seed for reproducible results.

print("Random numbers note:")
np.random.seed(42)

print("Random float in [0, 1):", np.random.rand())
print("Random integers from 1 to 9:", np.random.randint(1, 10))
print("Random sample of 5 values:\n", np.random.randint(0, 100, size=5))
```

```python
# More random number functions note:
# NumPy provides several random functions for different distributions and sampling tasks.

print("More random number functions note:")

print("Random numbers with rand(2, 3):\n", np.random.rand(2, 3))
print("Random sample with random_sample(5):\n", np.random.random_sample(5))
print("Standard normal values with randn(5):\n", np.random.randn(5))
print("Uniform values in [10, 20) with uniform(10, 20, size=5):\n", np.random.uniform(10, 20, size=5))
print("Normal values with normal(loc=0, scale=1, size=5):\n", np.random.normal(0, 1, size=5))
print("Binomial values with binomial(n=10, p=0.5, size=5):\n", np.random.binomial(10, 0.5, size=5))

items = np.array([10, 20, 30, 40, 50])
print("Random choice from array with choice(items, size=3, replace=False):\n", np.random.choice(items, size=3, replace=False))

shuffled_items = items.copy()
np.random.shuffle(shuffled_items)
print("Shuffled array with shuffle(items):\n", shuffled_items)

print("Random permutation with permutation(items):\n", np.random.permutation(items))
```
