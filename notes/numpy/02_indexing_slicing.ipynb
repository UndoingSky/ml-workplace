# Indexing and Slicing

Selections, indexing vs slicing, and examples.

```python
import numpy as np
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2, 3], [4, 5, 6]])

# Indexing vs slicing examples
print('Indexing example: b[0, 1] ->', b[0, 1])
print('Slicing example: b[:, 1] ->', b[:, 1])
print('Slicing example: a[1:3] ->', a[1:3])
```
