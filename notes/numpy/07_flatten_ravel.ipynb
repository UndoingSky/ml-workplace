# Flatten vs Ravel

Differences between `flatten()` and `ravel()` and examples showing when each returns a copy or a view.

```python
import numpy as np

# Example using flatten() -> returns a copy
original_arr = np.array([[1, 2], [3, 4]])
flat_arr = original_arr.flatten()
flat_arr[0] = 99
print('Original Array:', original_arr)
print('Flattened Array:', flat_arr)

# Example using ravel() -> returns a view when possible
original_arr = np.array([[1, 2], [3, 4]])
raveled_arr = original_arr.ravel()
raveled_arr[0] = 99
print('Original Array after ravel modification:', original_arr)
print('Raveled Array:', raveled_arr)

# Note: Both accept an `order` parameter ('C','F','A','K') to control traversal order.
```

**Behavior summary:**

- `flatten()` always returns a copy — modifying the result does not change the original array.
- `ravel()` returns a view when possible (no copy); modifying the result can modify the original array.
- Use `order` to control element traversal when flattening/raveling.
