# Arithmetic and Math Functions

Element-wise arithmetic, sqrt/round, and constants like pi.

```python
import numpy as np
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2, 3], [4, 5, 6]])

# Arithmetic operations
print('Addition: a + 2 ->', a + 2)
print('Subtraction: a - 1 ->', a - 1)
print('Multiplication: a * 3 ->', a * 3)
print('Division: a / 2 ->', a / 2)
print('Element-wise operations: b * 2 ->', b * 2)

# Math functions
print('Square root: np.sqrt(a) ->', np.sqrt(a))
print('Rounded sqrt: np.round(np.sqrt(a), 2) ->', np.round(np.sqrt(a), 2))

# Pi constant
print('Value of pi: np.pi ->', np.pi)
```
