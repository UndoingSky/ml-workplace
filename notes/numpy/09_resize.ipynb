# NumPy `resize` — Function vs Method

Notes and examples demonstrating the difference between `np.resize()` (function) and `arr.resize()` (method).

Key points:

- `np.resize(arr, shape)` returns a new array (copy). When expanding, it repeats the original data cyclically to fill the new shape.
- `arr.resize(shape)` is an array method that resizes in-place. When expanding, it pads with zeros where needed.
- Be careful: they have different behaviors and return/copy semantics.

```python
import numpy as np

arr = np.array([1, 2, 3, 4])

# Shrinking with the function -> cuts extra elements
shrunk = np.resize(arr, (2,))
print('Shrunk (np.resize):', shrunk)

# Expanding with the function -> repeats elements cyclically
expanded = np.resize(arr, (2, 4))
print('Expanded (np.resize):', expanded)
```

```python
# In-place resizing with the method -> pads with zeros when expanding
arr2 = np.array([1, 2, 3, 4])
arr2.resize((2, 3))
print('arr2 after arr2.resize((2,3)):', arr2)
```
