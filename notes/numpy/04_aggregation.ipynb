# Aggregation and Summary Statistics

Sum, mean, std, var, min and max (overall and along axes).

```python
import numpy as np
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2, 3], [4, 5, 6]])

# Aggregation examples
print('Sum: np.sum(a) ->', np.sum(a))
print('Mean: np.mean(a) ->', np.mean(a))
print('Standard deviation: np.std(a) ->', np.std(a))
print('Variance: np.var(a) ->', np.var(a))

print('Sum of b: np.sum(b) ->', np.sum(b))
print('Mean of b: np.mean(b) ->', np.mean(b))
print('Std of b: np.std(b) ->', np.std(b))

# Min / Max along axes
print('Min of b along axis 0 (columns):', np.min(b, axis=0))
print('Max of b along axis 1 (rows):', np.max(b, axis=1))
```
