# Transpose and Axis Reordering

Transposing 2D arrays and reordering axes for N-D arrays (useful for image/tensor manipulation).

```python
import numpy as np

# Create a 2x3 matrix
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Transpose it using .T
transposed = matrix.T

print('Original Shape:', matrix.shape)  # Output: (2, 3)
print('Transposed Shape:', transposed.shape) # Output: (3, 2)

print('
Transposed Matrix:
', transposed)
# Output:
# [[1 4]
#  [2 5]
#  [3 6]]
```

## Transposing 1D arrays

A 1D array has shape `(n,)` and `.T` has no effect. Convert to a 2D row vector before transposing.

```python
arr_1d = np.array([1, 2, 3])
print(arr_1d.shape)    # Output: (3,)
print(arr_1d.T.shape)  # Output: (3,) -- Nothing changed!

# Convert to a 2D row vector (shape: 1x3)
row_vector = arr_1d[np.newaxis, :]

# Now you can transpose it to a column vector (shape: 3x1)
col_vector = row_vector.T

print(col_vector)
# Output:
# [[1]
#  [2]
#  [3]]
```

## Advanced: N-D arrays and `np.transpose`

Use `np.transpose(arr, axes=...)` to specify a custom axis order rather than reversing all axes.

```python
# Shape: (2 batches, 10 height, 10 width, 3 color channels)
# The current axes indices are: (0, 1, 2, 3)
images = np.zeros((2, 10, 10, 3))

# Move Channels (axis 3) to be right after Batch (axis 0): new axes order = (0,3,1,2)
reshaped_images = np.transpose(images, axes=(0, 3, 1, 2))

print(reshaped_images.shape)  # Output: (2, 3, 10, 10)
```
