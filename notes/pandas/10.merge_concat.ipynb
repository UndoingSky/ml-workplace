# Merge and Concat## Main points to remember- merge joins tables by key columns or indexes.- concat stacks objects vertically or horizontally.- Choose the join type carefully: inner, left, right, or outer.- Check duplicate keys before merging to avoid unexpected row multiplication.## Quick reminderKeep the examples small and focus on the pandas API shape first, then the data-cleaning details.

```
import pandas as pdleft = pd.DataFrame({'id': [1, 2], 'name': ['Ana', 'Ben']})right = pd.DataFrame({'id': [1, 2], 'score': [91, 87]})merged = left.merge(right, on='id', how='left')combined = pd.concat([left, left], ignore_index=True)
```
