# Series and DataFrame## Main points to remember- A Series behaves like a labeled column.- A DataFrame combines multiple Series into one table.- Index values matter when aligning and comparing data.- Use vectorized operations instead of Python loops when possible.## Quick reminderKeep the examples small and focus on the pandas API shape first, then the data-cleaning details.

```
import pandas as pdscores = pd.Series({'Alice': 95, 'Bob': 88, 'Cara': 91})students = pd.DataFrame({'student': ['Alice', 'Bob', 'Cara'], 'score': [95, 88, 91]})print(scores['Alice'])print(students['score'].mean())
```
