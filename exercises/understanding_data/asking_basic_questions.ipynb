Basic questions before working on a dataset

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("titanic")
```

How big is the Data Set?

```python
df.shape  #tells the number of rows and columns. or How big is the dataset.
```

How does the data look like?

```python
df.head() #return's top 5 rows from the "df" dataset.  
```

df.head() gives the preview of dataset and helps in deciding which columns to to work on or which one to remove.

```python
df.sample(5) #return's 5 random rows from the "df" dataset
```

Whats is the datatype of columns?

```python
df.info()
```

```python
df = df.dropna(subset=["age"])
df["age"] = df["age"].astype(int)
df.info()
```

Are there any missing values?

```python
df.isnull().sum()
```

how does the data look mathematically?

```python
df.describe()
```

Are there any duplicate values?

```python
df.duplicated().sum()
```

How is the correlation between columns?

```python
df.corr()
```
