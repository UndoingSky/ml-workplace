# Combining tables with merge and concat

Real projects often split data across files or tables. Use `merge()` to join related tables and `concat()` to stack data together.

```python
import pandas as pd

students = pd.DataFrame({
    "student_id": [1, 2, 3],
    "name": ["Asha", "Ben", "Chen"],
})

scores = pd.DataFrame({
    "student_id": [1, 2, 3],
    "score": [72, 85, 90],
})

merged = students.merge(scores, on="student_id")
print(merged)

more_students = pd.DataFrame({
    "student_id": [4],
    "name": ["Dia"],
})

stacked = pd.concat([students, more_students], ignore_index=True)
print(stacked)
```

### Rule of thumb

- Use `merge()` when rows share a key column
- Use `concat()` when you want to stack rows or columns
- Joining data correctly is important for building ML datasets from multiple sources
