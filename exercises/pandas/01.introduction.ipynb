# Pandas topics you actually need for basic machine learning

Learn these first:

1. `Series` and `DataFrame` basics
2. Loading data with `read_csv()`
3. Inspecting data with `head()`, `info()`, `describe()`, `shape`, `columns`
4. Selecting data with `loc`, `iloc`, boolean filtering
5. Handling missing values with `isna()`, `fillna()`, `dropna()`
6. Cleaning column names and fixing data types with `astype()` and `to_datetime()`
7. Working with categorical data using `value_counts()` and `get_dummies()`
8. Creating or modifying columns for feature engineering
9. Sorting, filtering, and basic aggregation with `groupby()`
10. Combining tables with `merge()` and `concat()`
11. Exporting cleaned data with `to_csv()`

For starting ML, focus most on loading, cleaning, selecting features, handling missing values, and encoding categories. The rest can wait until you need it.
