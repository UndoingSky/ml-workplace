# Series and DataFrame basics

Use `Series` for one column of data and `DataFrame` for a table. These are the two core Pandas objects you will use everywhere in machine learning.

```python
import pandas as pd

# A Series is like a single labeled column
scores = pd.Series([72, 85, 90], name="score")
print(scores)

# A DataFrame is a table made of multiple columns
students = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen"],
    "age": [19, 21, 20],
    "score": [72, 85, 90],
})
print(students)

# Access one column as a Series
print(students["score"])

# Access a row by position
print(students.iloc[0])
```

### What to remember

- `Series` = one column
- `DataFrame` = table
- Most ML work starts by turning raw data into a clean DataFrame
