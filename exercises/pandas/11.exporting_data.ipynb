# Exporting cleaned data

After cleaning and feature engineering, you usually save the result so the next step in the ML pipeline can use it.

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Asha", "Ben"],
    "score": [72, 85],
    "passed": [True, True],
})

# Save to CSV for later use
df.to_csv("cleaned_students.csv", index=False)

# Read it back to confirm the output
loaded = pd.read_csv("cleaned_students.csv")
print(loaded)
```

### What to remember

- `to_csv()` is the most common beginner export format
- Set `index=False` when you do not want the row numbers saved as a column
- Saving clean data makes your ML workflow repeatable
