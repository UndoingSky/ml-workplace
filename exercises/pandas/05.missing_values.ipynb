# Handling missing values

Real datasets almost always have missing values. For basic ML, you need to detect them, decide whether to remove them, and often fill them with a reasonable value.

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dia"],
    "age": [19, None, 20, 18],
    "score": [72, 85, None, 65],
})

print(df.isna().sum())

# Fill numeric missing values with the column median
df["age"] = df["age"].fillna(df["age"].median())
df["score"] = df["score"].fillna(df["score"].median())

print(df)

# Drop rows if you want a smaller but fully complete dataset
clean_df = df.dropna()
print(clean_df)
```

### Practical rule

- Use `fillna()` when you want to keep the row
- Use `dropna()` when missing data is too much trouble or the dataset is large enough to lose some rows
- For ML, numeric columns are often filled with mean or median
