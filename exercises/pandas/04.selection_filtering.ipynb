# Selecting and filtering data

Selecting the right rows and columns is a core skill in Pandas. In ML, you use this to separate features from targets and to filter bad rows.

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Asha", "Ben", "Chen", "Dia"],
    "age": [19, 21, 20, 18],
    "score": [72, 85, 90, 65],
    "passed": [True, True, True, False],
})

# Select columns
print(df[["name", "score"]])

# Select rows by position
print(df.iloc[1:3])

# Select rows and columns by label
print(df.loc[df["score"] >= 80, ["name", "score"]])

# Boolean filtering
high_score = df[df["score"] >= 80]
print(high_score)
```

### What to remember

- Use `iloc` for position-based access
- Use `loc` for label-based access
- Boolean filters are how you keep only the rows you want
