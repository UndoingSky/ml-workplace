# Loading and inspecting data

Before training any model, load the dataset and inspect its shape, columns, and missing values. This tells you what kind of cleaning you need next.

```python
import pandas as pd
from io import StringIO

csv_data = StringIO("""name,age,score,passed
Asha,19,72,yes
Ben,21,85,yes
Chen,20,90,yes
Dia,18,,no
""")

df = pd.read_csv(csv_data)

# print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
print(df.isna().sum())
```

### Why this matters for ML

- `head()` shows the first rows
- `shape` tells you rows and columns
- `info()` shows data types and missing values
- `describe()` gives quick numeric stats
- `isna().sum()` shows how much cleanup is needed
