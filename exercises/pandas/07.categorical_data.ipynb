# Categorical data

Text columns with a small set of repeating values are categorical data. In ML, these usually need counting or encoding before you can use them in a model.

```python
import pandas as pd

df = pd.DataFrame({
    "city": ["Lagos", "Delhi", "Lagos", "Nairobi"],
    "grade": ["A", "B", "A", "C"],
})

print(df["city"].value_counts())

# One-hot encoding turns categories into numeric columns
encoded = pd.get_dummies(df, columns=["city", "grade"], drop_first=False)
print(encoded)

# If needed, keep a column as category type
df["city"] = df["city"].astype("category")
print(df.dtypes)
```

### What to remember

- `value_counts()` helps you understand category frequency
- `get_dummies()` is the common beginner-friendly encoding method
- Most models cannot use raw text categories directly
