# Feature engineering with Pandas

Feature engineering means creating useful input columns from raw data. For basic ML, this often matters more than the model choice itself.

```python
import pandas as pd

df = pd.DataFrame({
    "math": [70, 80, 90],
    "science": [65, 85, 95],
    "english": [75, 78, 88],
})

# Create a new feature from existing columns
df["average_score"] = df[["math", "science", "english"]].mean(axis=1)

# Create a binary feature
df["high_achiever"] = df["average_score"] >= 80

# Create a score gap feature
df["math_science_gap"] = df["math"] - df["science"]

print(df)
```

### Why this matters

- Models learn from features, not raw rows
- New columns can capture patterns that were not obvious before
- Simple features often improve a baseline model quickly
